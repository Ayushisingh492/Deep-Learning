```python
from sklearn.decomposition import TruncatedSVD
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
documents= ["""Sports and physical activities are essential for maintaining overall health and well-being. Regular exercise helps improve cardiovascular fitness, build muscle strength, and enhance mental health by reducing stress and anxiety. Different types of sports, such as team sports and individual activities, offer various benefits and opportunities for social interaction. Professional sports also play a significant role in entertainment and the global economy. Athletes often serve as role models, inspiring people to adopt healthier lifestyles. Despite the benefits, issues such as injuries, performance pressure, and doping in sports continue to be areas of concern."""]
```

# LSA

```python
vectorizer=TfidfVectorizer(stop_words='english')
x=vectorizer.fit_transform(documents)
```

```python
lsa=TruncatedSVD(n_components=3,random_state=42)
lsa_matrix=lsa.fit_transform(x)
```

```python
terms=vectorizer.get_feature_names_out()
for i, comp in enumerate(lsa.components_):
    print(f"\nTopic{i+1}:")
    terms_weights= zip(terms,comp)
    sorted_terms=sorted(terms_weights,key=lambda x:x[1],reverse=True)
    for term,weight in sorted_terms[:10]:
        print(term)
```

# NMF

```python
from sklearn.decomposition import NMF
```

```python
vectorizer=TfidfVectorizer(stop_words='english')
y=vectorizer.fit_transform(documents)
```

```python
#Apply NMF
nmf=NMF(n_components=2,random_state=42)
fiting=nmf.fit_transform(y)
hedding=nmf.components_
```

```python
#Display Topics
feature_names=vectorizer.get_feature_names_out()
for topic_idx,topic in enumerate(hedding):
  print(f"\nTopic {topic_idx+1}:")
  top_words=[feature_names[i] for i in topic.argsort()[-6:]]
  print(top_words)
```

# LDA

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
cv = CountVectorizer(max_df=1.0, min_df=1, stop_words='english')
z=cv.fit_transform(documents)
```

```python
LDA = LatentDirichletAllocation(n_components=4,random_state=42)
LDA.fit(z)
```

```python
for z,topic in enumerate(LDA.components_):
  print(f'the top4 words for topic {z}')
  print([cv.get_feature_names_out()[a] for a in topic.argsort()[-20:]])
  print('\n')
```
